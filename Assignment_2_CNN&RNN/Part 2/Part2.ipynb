```python
import matplotlib.pyplot as plt
def show_accu(x_axis1,y_axis1,x_axis2,y_axis2,y_axis3,x_label,y_label):
    # %matplotlib inline
    plt.plot(x_axis1,y_axis1,'g--')
    plt.plot(x_axis2,y_axis2,'r--')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    plt.plot(x_axis2,y_axis3,'b--')
    plt.xlabel(x_label)
    plt.ylabel("accuracy")
```

```python
show_accu(x_axis,loss_lst,x_axis_test,loss_lst_test,accu_lst,"sample","error")
```

#### 2. The accuracy and loss curves
Since the x-axis of every graph represents the number of input samples counting from the beginning of one "train" function
execution, we can get the approximate epoch number using this information. <br>
From the graphs, we shall see the test error first goes down in the first goes down with the training error going down in the
first three epochs. The best classification accuracy during this phase is 76%.
Then, with the training error continuously going down, the test error remains approximately
around 0.8 and 1 and the classification accuracy approximately 82%.<br>
#### 3. Discussion
From the graphs, we can get the above observation. Specifically, the test error remains decreasing
when the training samples are smaller than 200000. After that, the test error starts to increase and
enters the overfitting phase.<br>
During this phase, the training error will remain decreasing but the models starts to learn features that
are too specific to the training set. The model's complexity becomes too big to interpret the true data.<br>
Meanwhile, the classification accuracy does not appears too bad but stays unchanged, which corresponds
to the test error that has not changed much.
